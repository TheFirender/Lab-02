```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

array_size = (4, 4)
exponential_array = np.random.exponential(scale=1.0, size=array_size)

print(exponential_array)
```

```python

plt.figure(figsize=(12, 6))
exponential_array = np.random.exponential(scale=1.0, size=100000)
plt.hist(exponential_array, density=True, bins=50, histtype="step", color="red", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="blue", label="Uniform")
plt.hist(np.random.normal(loc=0, scale=1, size=100000), density=True, bins=50, histtype="step", color="green", label="Normal")

plt.axis([0, 5, 0, 1.2])
plt.legend(loc = "upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface Plot of Z = X^2 + Y^2')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

stats = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

pearson_corr = stats.corr(method='pearson')
spearman_corr = stats.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation")
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='YlGnBu')
plt.title("Spearman Rank Correlation")
plt.show()
```
