```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Au_nanoparticle_dataset.csv")

new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
new_df.head(20)
```

```python
print("Means:\n", new_df.mean())
print("\nStandard Deviations:\n", new_df.std())
print("\nQuartiles:\n", new_df.quantile([0.25, 0.5, 0.75]))
```

```python
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
plt.figure(figsize=(16, 4))

for i, feature in enumerate(features):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[feature], bins=30, color='skyblue', edgecolor='black')
    plt.title(feature)
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
g = sns.PairGrid(new_df)

g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot, fill=True)

plt.suptitle("Enhanced PairGrid with KDE", y=1.02)
plt.show()
```
